Filling missing values with a placeholder (Unknown)

```python
df = df.fillna("Unknown")
df.isnull().sum()
```

Adding 2 columns using ASIN (product_id), Brand name and product name(full name of the product)

```python
product_id_mapping = {
    'B085296FLT': {'brand': 'Beats', 'product': 'Beats Studio3'},
    'B0863TXGM3': {'brand': 'Sony', 'product': 'Sony WH-1000XM4 Wireless'},
    'B091FZK6VB': {'brand': 'JBL', 'product': 'JBL Live 660NC Wireless'},
    'B0B6GHW1SX': {'brand': 'Sennheiser', 'product': 'Sennheiser Momentum 4'},
    'B08FCGH2RL': {'brand': 'Skullcandy', 'product': 'Skullcandy Crusher Evo Wireless '},
    'B0CGTL9GFD': {'brand': 'Edifier', 'product': 'Edifier WH700NB Wireless'},
    'B09BYFHL25': {'brand': 'Audio-Technica', 'product': 'Audio-Technica ATH-M50xBT2 Wireless '},
    'B0CH1LT84X': {'brand': 'Bowers', 'product': 'Bowers & Wilkins Px7 S2e'},
    'B084R2WVLX': {'brand': 'Marshall', 'product': 'Marshall Monitor II Active '},
    'B09HL9MQ64': {'brand': 'Bose', 'product': 'Bose QuietComfort 45 Wireless'}
}

# Function to get brand name and product name using ASIN
def get_brand_product(asin):
    if asin in product_id_mapping:
        return product_id_mapping[asin]['brand'], product_id_mapping[asin]['product']
    else:
        return 'Unknown Brand', 'Unknown Product'

# Apply the function to the ASIN column and create new columns
df['brand_name'], df['product_name'] = zip(*df['product_id'].apply(get_brand_product))

df.head()
```

Adding a column for date (extracting from timestamp)

```python
# Function to extract the date from the timestamp
def extract_date(timestamp):
    # Split the string and take the last part (the date)
    try:
        date_part = timestamp.split("Reviewed in the United States ")[-1]
        return pd.to_datetime(date_part, format='%B %d, %Y', errors='coerce')
    except Exception as e:
        return pd.NaT  # Return Not-a-Time for any errors

# Create a new column 'date' to hold the extracted dates
df['date'] = df['timestamp'].apply(extract_date)
df.head()
```

Adding a column of content length, which is the length of the review.

```python
# Calculate the length of content
df['content_length'] = df['content'].apply(len)
```

This scatter plot helps us identify if longer reviews tend to give more positive or negative ratings. There might be significant relationship here so checking using Pearson correlation.

```python
# Calculate the Pearson correlation
correlation, p_value = pearsonr(df['content_length'], df['rating'])

# Print the correlation coefficient and p-value
print(f'Pearson correlation coefficient: {correlation:.3f}')
print(f'P-value: {p_value:.3f}')

# Determine the relationship and significance
if p_value < 0.05:
    significance = "statistically significant"
else:
    significance = "not statistically significant"

if correlation > 0:
    relationship = "positive"
elif correlation < 0:
    relationship = "negative"
else:
    relationship = "no"

print(f'The relationship between review length and rating is {relationship} and is {significance}.')
```

The value of -0.107 suggests a very weak negative relationship. This means that, on average, slightly longer reviews tend to have slightly lower ratings, but the relationship is not strong.

The p-value of 0.001 is significantly less than 0.05, indicating that the observed correlation is statistically significant.

Overall, the results suggest that there's a very weak tendency for longer reviews to have slightly lower ratings.

### 3.3 Correlation Analysis

**Key Insights from Correlation and P-Value Matrices**

- **Correlation between Content Length and Helpful Count**
A positive correlation of **0.37** exists between `content_length` and `helpful_count`. This indicates that longer reviews tend to receive more helpful votes, though this correlation is moderate.

- **Weak Correlations with Rating**
The `rating` variable shows weak correlations with both `helpful_count` (-0.05) and `content_length` (-0.11). This suggests that the length of a review and how helpful it is perceived have minimal influence on the product's rating.

- **P-Values Indicate Significant Relationships**
Based on the P-Value Matrix, the relationships between `content_length` and `helpful_count`, as well as `helpful_count` and `rating`, are statistically significant with p-values close to **0.00**, which are well below the significance threshold of **0.05**.

- **Statistically Insignificant Correlations for Rating**
The correlation between `rating` and both `content_length` and `helpful_count` is not statistically significant, as indicated by the p-value of **0.11**. This suggests that variations in review length or helpful votes do not significantly impact the product ratings.

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Initialize a matrix for p-values
p_value_matrix = np.zeros(correlation_matrix.shape)

# Calculate p-values for each pair
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        if i != j:
            _, p_value = pearsonr(numeric_df.iloc[:, i], numeric_df.iloc[:, j])
            p_value_matrix[i, j] = p_value
        else:
            p_value_matrix[i, j] = np.nan  # Set diagonal to NaN (no correlation)

# Convert p-value matrix to a DataFrame for better visualization
p_value_df = pd.DataFrame(p_value_matrix, index=correlation_matrix.columns, columns=correlation_matrix.columns)

# Set up the matplotlib figure for correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar=True)
plt.title('Correlation Matrix')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

# Set up the matplotlib figure for p-value heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(p_value_df, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar=True, mask=np.isnan(p_value_df))
plt.title('P-Value Matrix')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

# Interpretation of significance
significance_threshold = 0.05
significant_correlations = p_value_df < significance_threshold
print("Significant correlations (p < 0.05):")
print(significant_correlations)
```

```python

from nltk.sentiment.vader import SentimentIntensityAnalyzer


# TextBlob Sentiment Analysis
def get_textblob_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

# Applying TextBlob on reviews
df['polarity'], df['subjectivity'] = zip(*df['content'].apply(get_textblob_sentiment))

# VADER Sentiment Analysis
sia = SentimentIntensityAnalyzer()

def get_vader_sentiment(text):
    return sia.polarity_scores(text)

# Applying VADER on reviews
df['vader_sentiment'] = df['content'].apply(get_vader_sentiment)
df['vader_compound'] = df['vader_sentiment'].apply(lambda x: x['compound'])

# Visualizing sentiment distribution using histograms
plt.figure(figsize=(12, 6))

# Polarity distribution from TextBlob
plt.subplot(1, 2, 1)
plt.hist(df['polarity'], bins=20, color='blue', alpha=0.7)
plt.title('TextBlob Sentiment Polarity Distribution')
plt.xlabel('Polarity')
plt.ylabel('Frequency')

# Compound score distribution from VADER
plt.subplot(1, 2, 2)
plt.hist(df['vader_compound'], bins=20, color='green', alpha=0.7)
plt.title('VADER Sentiment Compound Score Distribution')
plt.xlabel('Compound Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

- There are fewer negative reviews, as seen in the low frequency of polarity values below **0**.

**VADER Sentiment Compound Score Distribution**

```python
def get_sentiment(review):
    analysis = TextBlob(review)
    # Classify the sentiment
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df['content'].apply(get_sentiment)
sentiment_counts = df['sentiment'].value_counts()
print(sentiment_counts)
sentiment_summary = df.groupby(['product_name', 'sentiment']).size().unstack(fill_value=0)

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a grouped bar chart
sentiment_summary.plot(kind='bar', stacked=False, color=['lightgreen', 'salmon', 'lightgrey'], edgecolor='black')

plt.title('Sentiment Analysis by Product')
plt.xlabel('Product Name')
plt.ylabel('Count of Sentiments')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```

**Model Setup**

```python
api_key = "Insert"

# Function to analyze reviews using Gemini AI
def analyze_reviews(reviews, api_key, prompt_message):
    genai.configure(api_key=api_key)

    # Define generation configuration directly
    generation_config = {
        "temperature": 1,
        "top_p": 0.95,
        "top_k": 64,
        "max_output_tokens": 8192,
        "response_mime_type": "text/plain",
    }

    model = genai.GenerativeModel(
        model_name="gemini-1.5-flash",
        generation_config=generation_config,
    )

    chat_session = model.start_chat(history=[])

    prompt = prompt_message + "\n\n".join(reviews)
    response = chat_session.send_message(prompt)
    return response.text
```

**Prompt 1: Identify Common Themes or Topics in Reviews**

```python
# Define the prompt message for identifying common themes
prompt_message_1 = (
    "Analyze the following product reviews and identify the common themes or topics discussed:\\n\\n"
)

# Select all reviews' content
all_reviews = df['content'].tolist()

# Call the analyze_reviews function
common_themes = analyze_reviews(all_reviews, api_key, prompt_message_1)

# Display the response
print("Common Themes or Topics in Reviews:")
print(common_themes)
```
